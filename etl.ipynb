```python
import requests
import json
import oracledb
import re
from typing import Dict, List, Any, Optional
from concurrent.futures import ThreadPoolExecutor
import logging

# Configure logging
logging.basicConfig(
    level=logging.INFO, 
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.StreamHandler(),
        logging.FileHandler('fhir_processing.log')
    ]
)
logger = logging.getLogger(__name__)

# === CONFIGURATION ===
CONFIG = {
    'base_par_url': "https://objectstorage.ap-batam-1.oraclecloud.com/p/i01zKdA6zq1WivKf0iDFfvBcMOpOeEScqhdafJJq7_PF4cX_Q2wPY7mMfP-d_hRN/n/axk5suia05qz/b/sample-FHIR-syntetic-canadian/o/",
    'prefix': "",
    'db_config': {
        'user': "ADMIN",
        'password': "OcsAIEngineer2045MoH",
        'dsn': "rpadb05_high",
        'config_dir': "Wallet_RPADB05",
        'wallet_location': "Wallet_RPADB05",
        'wallet_password': "OcsAIEngineer2045MoH"
    },
    'batch_size': 1000,
    'max_workers': 5
}

# === UTILITY FUNCTIONS ===
def strip_timezone(ts: Optional[str]) -> Optional[str]:
    """Remove timezone from timestamp string"""
    if ts and isinstance(ts, str):
        return re.sub(r"[+-]\d{2}:\d{2}$", "", ts)
    return ts

from datetime import datetime
def normalize_birth_date(date_str):
    if isinstance(date_str, str) and re.match(r"\d{4}-\d{2}-\d{2}", date_str):
        return date_str  # format valid
    return None

def safe_get(data: Dict, *keys, default=None) -> Any:
    """Safely get nested dictionary values"""
    for key in keys:
        if isinstance(data, dict) and key in data:
            data = data[key]
        elif isinstance(data, list) and isinstance(key, int) and 0 <= key < len(data):
            data = data[key]
        else:
            return default
    return data

def get_reference_id(reference: Optional[str]) -> Optional[str]:
    """Extract ID from FHIR reference string"""
    if reference and isinstance(reference, str):
        return reference.split('/')[-1]
    return reference

# === SQL STATEMENTS ===
SQL_STATEMENTS = {
    'condition': """
        INSERT INTO CONDITION (
            ID, CLINICAL_STATUS, VERIFICATION_STATUS, CODE, TEXT,
            ENCOUNTER_ID, ONSET_TIME, ABATEMENT_TIME, RECORDED_TIME
        ) VALUES (
            :id, :clinical_status, :verification_status, :code, :text,
            :encounter_id,
            TO_TIMESTAMP(:onset_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            TO_TIMESTAMP(:abatement_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            TO_TIMESTAMP(:recorded_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF')
        )
    """,
    'encounter': """
        INSERT INTO ENCOUNTER (
            ID, STATUS, TYPE, PATIENT_ID,
            PARTICIPANT_TYPE, PARTICIPANT_NAME,
            START_TIME, END_TIME, SERVICE_PROVIDER
        ) VALUES (
            :id, :status, :type, :patient_id,
            :participant_type, :participant_name,
            TO_TIMESTAMP(:start_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            TO_TIMESTAMP(:end_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            :service_provider
        )
    """,
    'procedure': """
        INSERT INTO MEDICAL_PROCEDURE (
            ID, STATUS, PATIENT_ID, ENCOUNTER_ID,
            CODE, TEXT, PERIOD_START, PERIOD_END
        ) VALUES (
            :id, :status, :patient_id, :encounter_id,
            :code, :text, TO_TIMESTAMP(:period_start, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            TO_TIMESTAMP(:period_end, 'YYYY-MM-DD"T"HH24:MI:SS.FF')
        )
    """,
    'observation': """
        INSERT INTO OBSERVATION (
            ID, STATUS, PATIENT_ID, ENCOUNTER_ID,
            CATEGORY, CODE, TEXT, VALUE, UNIT,
            EFFECTIVE_TIME, ISSUED_TIME
        ) VALUES (
            :id, :status, :patient_id, :encounter_id,
            :category, :code, :text, :value, :unit,
            TO_TIMESTAMP(:effective_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF'),
            TO_TIMESTAMP(:issued_time, 'YYYY-MM-DD"T"HH24:MI:SS.FF')
        )
    """,
    'patient': """
        INSERT INTO PATIENT (
            ID, NAME_FAMILY, NAME_GIVEN, GENDER,
            PHONE_NUMBER, BIRTH_DATE, CITY,
            STATE, POSTAL_CODE, COUNTRY,
            MARITAL_STATUS, LANGUAGE
        ) VALUES (
            :id, :name_family, :name_given, :gender,
            :phone_number, TO_DATE(:birth_date, 'YYYY-MM-DD'), :city,
            :state, :postal_code, :country,
            :marital_status, :language
        )
    """
}

# === RESOURCE PROCESSORS ===
def process_condition(resource: Dict) -> Dict:
    """Process Condition resource"""
    return {
        "id": resource.get("id"),
        "clinical_status": safe_get(resource, "clinicalStatus", "coding", 0, "code"),
        "verification_status": safe_get(resource, "verificationStatus", "coding", 0, "code"),
        "code": safe_get(resource, "code", "coding", 0, "code"),
        "text": safe_get(resource, "code", "coding", 0, "display"),
        "patient_id": get_reference_id(safe_get(resource, "subject", "reference")),
        "encounter_id": get_reference_id(safe_get(resource, "encounter", "reference")),
        "onset_time": strip_timezone(resource.get("onsetDateTime")),
        "abatement_time": strip_timezone(resource.get("abatementDateTime")),
        "recorded_time": strip_timezone(resource.get("recordedDate"))
    }

def process_encounter(resource: Dict) -> Dict:
    """Process Encounter resource"""
    participant = safe_get(resource, "participant", 0, default={})
    period = participant.get("period", {})
    
    return {
        "id": resource.get("id"),
        "status": resource.get("status"),
        "type": safe_get(resource, "type", 0, "text"),
        "patient_id": get_reference_id(safe_get(resource, "subject", "reference")),
        "participant_type": safe_get(participant, "type", 0, "text"),
        "participant_name": safe_get(participant, "individual", "display"),
        "start_time": strip_timezone(period.get("start")),
        "end_time": strip_timezone(period.get("end")),
        "service_provider": safe_get(resource, "serviceProvider", "display")
    }

def process_observation(resource: Dict) -> Dict:
    """Process Observation resource"""
    return {
        "id": resource.get("id"),
        "status": resource.get("status"),
        "patient_id": get_reference_id(safe_get(resource, "subject", "reference")),
        "encounter_id": get_reference_id(safe_get(resource, "encounter", "reference")),
        "category": safe_get(resource, "category", 0, "coding", 0, "display"),
        "code": safe_get(resource, "code", "coding", 0, "code"),
        "text": safe_get(resource, "code", "coding", 0, "display"),
        "value": safe_get(resource, "valueQuantity", "value"),
        "unit": safe_get(resource, "valueQuantity", "unit"),
        "effective_time": strip_timezone(resource.get("effectiveDateTime")),
        "issued_time": strip_timezone(resource.get("issued"))
    }

def process_patient(resource: Dict) -> Dict:
    """Process Patient resource"""
    name = safe_get(resource, "name", 0, default={})
    telecom = safe_get(resource, "telecom", 0, default={})
    address = safe_get(resource, "address", 0, default={})
    comm = safe_get(resource, "communication", 0, default={})
    
    return {
        "id": resource.get("id"),
        "name_family": name.get("family"),
        "name_given": safe_get(name, "given", 0),
        "gender": resource.get("gender"),
        "phone_number": telecom.get("value"),
        "birth_date": normalize_birth_date(resource.get("birthDate", None)),
        "city": address.get("city"),
        "state": address.get("state"),
        "postal_code": address.get("postalCode"),
        "country": address.get("country"),
        "marital_status": safe_get(resource, "maritalStatus", "text"),
        "language": safe_get(comm, "language", "text")
    }

def process_procedure(resource: Dict) -> Dict:
    """Process Procedure resource"""
    return {
        "id": resource.get("id"),
        "status": resource.get("status"),
        "patient_id": get_reference_id(safe_get(resource, "subject", "reference")),
        "encounter_id": get_reference_id(safe_get(resource, "encounter", "reference")),
        "code": safe_get(resource, "code", "coding", 0, "code"),
        "text": safe_get(resource, "code", "coding", 0, "display"),
        "period_start": strip_timezone(safe_get(resource, "performedPeriod", "start")),
        "period_end": strip_timezone(safe_get(resource, "performedPeriod", "end"))
    }

# === RESOURCE PROCESSOR MAPPING ===
RESOURCE_PROCESSORS = {
    "Condition": (process_condition, "condition"),
    "Encounter": (process_encounter, "encounter"),
    "Observation": (process_observation, "observation"),
    "Patient": (process_patient, "patient"),
    "Procedure": (process_procedure, "procedure")
}

# === MAIN FUNCTIONS ===
def get_json_files() -> List[str]:
    """Get list of JSON files from object storage"""
    list_url = CONFIG['base_par_url'] + "?prefix=" + CONFIG['prefix']
    
    try:
        response = requests.get(list_url, timeout=30)
        response.raise_for_status()
        
        objects = response.json().get("objects", [])
        files = [obj["name"] for obj in objects if obj["name"].endswith(".json")]
        
        logger.info(f"Found {len(files)} JSON files")
        return files
        
    except requests.RequestException as e:
        logger.error(f"Failed to get file list: {e}")
        return []

def process_json_file(file_name: str) -> List[Dict]:
    """Process a single JSON file and return processed records - STOP on error"""
    file_url = CONFIG['base_par_url'] + requests.utils.quote(file_name, safe="")
    
    try:
        response = requests.get(file_url, timeout=60)
        response.raise_for_status()
        
        data = response.json()
        processed_records = []
        
        for entry in data.get("entry", []):
            resource = entry.get("resource", {})
            resource_type = resource.get("resourceType")
            
            if resource_type in RESOURCE_PROCESSORS:
                processor_func, table_name = RESOURCE_PROCESSORS[resource_type]
                try:
                    processed_data = processor_func(resource)
                    processed_records.append((table_name, processed_data))
                except Exception as e:
                    resource_id = resource.get("id", "unknown")
                    logger.error(f" FATAL ERROR - Processing {resource_type} failed!")
                    logger.error(f"   File: {file_name}")
                    logger.error(f"   Resource ID: {resource_id}")
                    logger.error(f"   Error: {e}")
                    logger.error(f"   Resource data: {resource}")
                    
                    # Raise exception to stop processing
                    raise Exception(f"Processing {resource_type} failed in file {file_name} for resource {resource_id}: {e}")
        
        logger.info(f" Successfully processed {file_name}: {len(processed_records)} records")
        return processed_records
        
    except requests.RequestException as e:
        logger.error(f" FATAL ERROR - Failed to fetch {file_name}: {e}")
        raise Exception(f"Failed to fetch file {file_name}: {e}")
    except json.JSONDecodeError as e:
        logger.error(f" FATAL ERROR - Invalid JSON in {file_name}: {e}")
        raise Exception(f"Invalid JSON in file {file_name}: {e}")
    except Exception as e:
        logger.error(f" FATAL ERROR - Unexpected error processing {file_name}: {e}")
        raise  # Re-raise the exception to stop processing

def batch_insert_records(cursor, records: List[tuple]) -> None:
    """Insert records in batches grouped by table - STOP on first error"""
    # Group records by table
    table_records = {}
    for table_name, record_data in records:
        if table_name not in table_records:
            table_records[table_name] = []
        table_records[table_name].append(record_data)
    
    # Insert records by table in batches
    for table_name, record_list in table_records.items():
        if table_name in SQL_STATEMENTS:
            sql = SQL_STATEMENTS[table_name]
            
            # Process in batches
            for i in range(0, len(record_list), CONFIG['batch_size']):
                batch = record_list[i:i + CONFIG['batch_size']]
                try:
                    cursor.executemany(sql, batch)
                    logger.info(f" Inserted batch of {len(batch)} records into {table_name.upper()}")
                except Exception as e:
                    logger.error(f" BATCH INSERT FAILED into {table_name.upper()}: {e}")
                    logger.info(f" Identifying problematic record...")
                    
                    # Try individual inserts to find the exact problematic record
                    for idx, record in enumerate(batch):
                        try:
                            cursor.execute(sql, record)
                            logger.debug(f" Record {idx + 1}/{len(batch)} inserted successfully")
                        except Exception as individual_error:
                            record_id = record.get('id', f'batch_{i}_index_{idx}')
                            logger.error(f" FATAL ERROR - INSERT FAILED!")
                            logger.error(f"   Table: {table_name.upper()}")
                            logger.error(f"   Record ID: {record_id}")
                            logger.error(f"   Error: {individual_error}")
                            logger.error(f"   Record data: {record}")
                            logger.error(f"   SQL: {sql}")
                            
                            # Raise exception to stop processing
                            raise Exception(f"INSERT FAILED into {table_name.upper()} for record {record_id}: {individual_error}")
                    
                    # If we get here, the batch error was temporary, continue
                    logger.info(f" All individual records in batch processed successfully")

def main():
    """Main processing function"""
    logger.info("Starting FHIR data processing...")
    
    # Get JSON files
    files = get_json_files()
    if not files:
        logger.error(" No files found to process")
        return
    
    # Connect to database
    try:
        connection = oracledb.connect(**CONFIG['db_config'])
        cursor = connection.cursor()
        logger.info(" Connected to database")
    except Exception as e:
        logger.error(f" Database connection failed: {e}")
        return
    
    # Statistics tracking
    stats = {
        'total_files': len(files),
        'processed_files': 0,
        'total_records': 0,
        'records_by_table': {}
    }
    
    try:
        # Process files in parallel
        all_records = []
        with ThreadPoolExecutor(max_workers=CONFIG['max_workers']) as executor:
            # Submit all tasks
            future_to_file = {executor.submit(process_json_file, file): file for file in files}
            
            # Process results as they complete
            for future in future_to_file:
                file_name = future_to_file[future]
                try:
                    file_records = future.result()  # This will raise exception if processing failed
                    if file_records:
                        stats['processed_files'] += 1
                        all_records.extend(file_records)
                        
                        # Count records by table
                        for table_name, _ in file_records:
                            stats['records_by_table'][table_name] = stats['records_by_table'].get(table_name, 0) + 1
                    
                except Exception as e:
                    logger.error(f" PROCESSING STOPPED due to error in file: {file_name}")
                    logger.error(f"   Error: {e}")
                    
                    # Cancel all remaining tasks
                    for remaining_future in future_to_file:
                        remaining_future.cancel()
                    
                    # Rollback and exit
                    logger.info(" Rolling back database changes...")
                    connection.rollback()
                    raise Exception(f"Processing stopped due to error in file {file_name}: {e}")
        
        stats['total_records'] = len(all_records)
        
        # Log processing summary
        logger.info(f" PROCESSING SUMMARY:")
        logger.info(f"   Total files: {stats['total_files']}")
        logger.info(f"    Processed: {stats['processed_files']}")
        logger.info(f"    Total records: {stats['total_records']}")
        
        for table_name, count in stats['records_by_table'].items():
            logger.info(f"    {table_name.upper()}: {count} records")
        
        # Insert all records
        if all_records:
            logger.info(" Starting database insertions...")
            batch_insert_records(cursor, all_records)  # This will stop on first error
            connection.commit()
            logger.info(" All records committed to database")
        else:
            logger.warning("  No records to insert")
        
    except Exception as e:
        logger.error(f" PROCESSING FAILED: {e}")
        logger.info(" Rolling back database changes...")
        connection.rollback()
        raise  # Re-raise to ensure the script exits with error code
        
    finally:
        cursor.close()
        connection.close()
        logger.info(" Database connection closed")
    
    logger.info(" FHIR data processing completed successfully")

if __name__ == "__main__":
    main()
```
